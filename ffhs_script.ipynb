# Transform files to Pandas DataFrame

```python
df_immobles = pd.read_excel('/content/1_FFHS CONSULTA PRINEX IMMOBLES_v2.xlsx',sheet_name="CSV")
df_crm = pd.read_excel('/content/2_FFHS CONSULTA PRINEX CRM.xlsx',sheet_name="CSV_01")
df_immobles_clients = pd.read_excel('/content/3_FFHS CONSULTA PRINEX IMMOBLES I CLIENTS.xlsx',sheet_name='D4G')
df_fact = pd.read_excel('/content/4_FFHS_FACT_2016_2019_ANUAL.xlsx',sheet_name='D4G')
df_morositat = pd.read_excel('/content/5_MOROSITAT_FFHS_2016_2019_FINAL.xlsx',sheet_name='Hoja1')
df_ajuts_sae = pd.read_excel('/content/6_AJUTS_SAE_2016_2019.xlsx',sheet_name='D4G')
df_ajuts_ahc = pd.read_excel('/content/7_AJUTS AHC_2016_2019.xlsx',sheet_name='D4G')
df_info_fiscal = pd.read_excel('/content/8_FFHS_INFO_SOCIAL_2016_2019_feb_2021.xlsx',sheet_name='Ad4good nov 2020')
```

```python
#Step 1) Changing the Index of the Immobles and Clients Dataframes prior to execute the merge
df_immobles.set_index('Id_Inmueble Prinex',inplace=True)
df_immobles_clients.set_index('Inmueble',inplace=True)
```

```python
df_merged_immobles_clients = pd.merge(df_immobles,df_immobles_clients,how='outer',left_on='Id_Inmueble Prinex',right_on='Inmueble')
df_merged_immobles_clients.head()
```
